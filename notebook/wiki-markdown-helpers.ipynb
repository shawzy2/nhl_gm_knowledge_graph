```python
h = sorted(list(hockey_ops))
c = sorted(list(coaching))
o = sorted(list(ownership))
sc = sorted(list(scouting))
su = sorted(list(support))

len_longest_list = -1
for li in [h, c, o, sc, su]:
    if len(li) > len_longest_list:
        len_longest_list = len(li)
```

```python
for i in range(0, len_longest_list):
    print('| ', end='')
    if len(h) > i: print(h[i], end='')
    print('  | ', end='')
    if len(c) > i: print(c[i], end='')
    print('  | ', end='')
    if len(o) > i: print(o[i], end='')
    print('  | ', end='')
    if len(sc) > i: print(sc[i], end='')
    print('  | ', end='')
    if len(su) > i: print(su[i], end='')
    print('  | ', end='')
    
    print()
```
