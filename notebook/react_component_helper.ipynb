```python
import os
import pandas as pd
import numpy as np
```

```python
path = os.getcwd()
path = os.path.abspath(os.path.join(path, os.pardir, 'data/gm_history.csv'))
with open(path) as f:
    df_gm_hist = pd.read_csv(f)
```

```python
df_gm_hist['end_date'] = df_gm_hist['end_date'].str.strip()
df_gm_hist['end_date'] = df_gm_hist['end_date'].str.lower()

df_gm_hist['team'] = df_gm_hist['team'].str.strip()
df_gm_hist['team'] = df_gm_hist['team'].str.lower()
```

```python
df_gm_hist[df_gm_hist['end_date']=='present']
```

```python
teams = list(df_gm_hist[df_gm_hist['end_date']=='present']['team'])
for team in teams:
    team_str = team.split()[0]
    if team == 'new york rangers' or team == 'new york islanders':
        team_str = team.replace(' ', '')
    print("import { ReactComponent as " + team_str.capitalize() + " } from './logos/" + team_str + ".svg';")
```

```python
present_gms = list(df_gm_hist[df_gm_hist['end_date']=='present']['name'])
teams = list(df_gm_hist[df_gm_hist['end_date']=='present']['team'])

for i in range(len(present_gms)):
    print('<DropdownItem onClick={() => setGeneralManager("' + present_gms[i] + '")} leftIcon={<' + teams[i].split()[0].capitalize() + ' />} >' + present_gms[i] + '</DropdownItem>')
```
