```python
import requests
import pprint
import os
import json
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd

import ssl

ssl._create_default_https_context = ssl._create_unverified_context
```

### Parse list of NHL/AHL team links

```python
# get list of links to NHL team pages
team_and_links = {}

try:
    url = 'https://www.eliteprospects.com/league/nhl'
    page = urlopen(url)
    html = page.read().decode("utf-8")
    soup = BeautifulSoup(html, 'html.parser')
    
    teams_tags = soup.find('div', {'class': 'leg-home-inner'}).find_all('a')
    for tag in teams_tags:
        team_name = tag.get_text().strip()
        team_link = tag.get('href')
        team_and_links[team_name] = team_link
        
except Exception as e:
    print('failed on page ' + url)
    print(e)
```

```python
for team_name, team_link in team_and_links.items():
    print(team_name)
    print(team_link + '/depth-chart')
```

```python
player_tags = soup.find('table', {'class': 'table table-striped table-sortable depth-chart highlight-stats'}) \
    .find_all("tr") 
```

```python
players = []
for tag in player_tags:
    if tag.find('a') is not None:
        # extract details from row in table
        name_and_position = tag.find('a').get_text().strip()
        name = name_and_position[:name_and_position.find('(')-1]
        position = name_and_position[name_and_position.find('(')+1:name_and_position.find('/')]
        link = tag.find('a').get('href')
        birth_year = tag.find('td', {'class': 'date-of-birth'}).get_text().strip()
        contract = tag.find('td', {'class': 'contract'}).get_text().strip()
        cap_hit = tag.find('td', {'class': 'cap-hit'}).get_text().strip()
        seasons = tag.find('td', {'class': 'seasons'}).get_text().strip()
        gp = tag.find('td', {'class': 'gp'}).get_text().strip()
        g = tag.find('td', {'class': 'g'}).get_text().strip()
        a = tag.find('td', {'class': 'a'}).get_text().strip()
        tp = tag.find('td', {'class': 'tp'}).get_text().strip()
        pim = tag.find('td', {'class': 'pim'}).get_text().strip()
        plusminus = tag.find('td', {'class': 'pm'}).get_text().strip()
        
        # shove details into row and add row to table
        row = [
            name,
            position,
            birth_year,
            contract,
            cap_hit,
            seasons,
            gp,
            g,
            a,
            tp,
            pim,
            plusminus,
            link
        ]
        players.append(row)
```
