```python
import json
import pandas as pd
import pprint
```

### Read in data

```python
# read file containing staff and positions
with open('active_staff_all_time_history.json') as f:
    active_staff_all_time_history = json.load(f)
    
print(len(active_staff_all_time_history))
```

### Convert to DF

```python
df = pd.DataFrame(active_staff_all_time_history)
df.columns = ['name', 'season', 'team', 'league', 'position', 'notes', 'image']
print(str(len(df['name'].unique())) + ' distinct staff members')
```

#### Use Case: Percentage of Staff from Past Relationships
find how many of Kyle Dubas' staff has worked with Dubas in the past on different teams

```python
df[df['name']=='Kyle Dubas']
```

```python
name = 'Kyle Dubas'
df[df['name']==name]
history = [tuple(r) for r in df[df['name']==name].to_numpy()]

relationships = []
for row in history:
    season = row[1]
    team = row[2]
    df_tmp = df[(df['name']!=name) \
                      & (df['season']==season) \
                      & (df['team']==team)]
    relationships.extend([tuple(r) for r in df_tmp.to_numpy()])
```

```python
season_and_coworkers = {}
name = 'Ken Holland'
current_season = '2020-21'
current_team = df[(df['name']==name) * (df['league']=='NHL')].sort_values(by='season')['team'].to_list()[-1]
print(name, current_season, current_team, sep='\n')

# find prev relationships
for index, row in df[df['name']==name].iterrows():
    season = row['season']
    team = row['team']
    
    season_and_coworkers[season] = df[(df['name']!=name) \
                                      & (df['season']==season) \
                                      & (df['team']!=current_team) \
                                      & (df['team']==team)]['name'].tolist()


# get set of current staff members
current_staff = set(df[(df['team']==current_team) & (df['season']==current_season)]['name'].to_list())

```

```python
season_and_coworkers
```

```python
hired_from_prev_relationship = []
for season, staff_list in season_and_coworkers.items():
#     print(str(season) + ':')
    for staff_member in staff_list:
        if staff_member in current_staff:
#             print(staff_member)
            display(df[(df['name']==staff_member) & (df['season']==season)])
            team_this_season = df[(df['name']==name) & (df['season']==season)]['team'].to_list()[-1]
            league_this_season = df[(df['name']==name) & (df['season']==season)]['league'].to_list()[-1]
            position_this_season = df[(df['name']==staff_member) & (df['season']==season)]['position'].to_list()[-1]
            hired_from_prev_relationship.append([staff_member, season, team_this_season, league_this_season, position_this_season])
#     print()
```

```python
display(pd.DataFrame(hired_from_prev_relationship).sort_values(by=[0,1]))
```
